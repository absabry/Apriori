```python
from utils import dataset_from_file,apriori,format_output
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('darkgrid')
```

```python
delimeter = " "
dataset = "datasets/dataset_words.txt"
T = dataset_from_file(dataset, delimeter)
eps = 2
verbose = False
```

```python
res = apriori(T,eps,verbose)
format_output(res)
```

```python
plt.figure(figsize=(20,8))
ax = sns.barplot(
    x=[x[0] for x in res], 
    y=[x[1] for x in res], 
    palette="Blues_d"); 
ax.set_xticklabels(ax.get_xticklabels(), fontsize=15)
ax.set_ylabel(ax.get_ylabel(), fontsize=15)
ax.set_title('Terms frequency', fontsize=20)
ax.set_ylabel('Occurence', fontsize=15)
ax.set_xlabel('Terms', fontsize=15); 
```
